```python
file_path = '/content/drive/My Drive/Intro to ML/Datasets/D3.csv'
df = pd.DataFrame(pd.read_csv(file_path))
```

```python
x1 = df.values[: , 0] # Import X1 (all rows in first column)
x2 = df.values[: , 1] # Import X2 (all rows in second column)
x3 = df.values[: , 2] # Import X3 (all rows in third column)
Y = df.values[: , 3] # Import outout (all rows in fourth column)

m = len(x1)
```

```python
# Create ones matrix & reshape others for matrix operation
ones = np.ones((m,1))
X_1 = x1.reshape(m,1)
X_2 = x2.reshape(m,1)
X_3 = x3.reshape(m,1)

# Stack matrices horizontally
X1 = np.hstack((ones, X_1))
X2 = np.hstack((ones, X_2))
X3 = np.hstack((ones, X_3))
```

```python
theta1 = np.zeros(2)
theta2 = np.zeros(2)
theta3 = np.zeros(2)
```

```python
def compute_cost(X, Y, theta):
  '''
  X : 2D array training samples
  Y : 1D array target values
  Theta: parameters
  '''
  # Multiply X matrix with Theta matrix (hypothesis function)
  predictions = X.dot(theta)
  # Perform [1/(2m)(h-y)^2] to find loss function
  errors = np.subtract(predictions,Y)
  sqrErrors = np.square(errors)
  J = 1/(2*m) * np.sum(sqrErrors)
  return J
```

```python
costX1 = compute_cost(X1,Y,theta1)
costX2 = compute_cost(X2,Y,theta2)
costX3 = compute_cost(X3,Y,theta3)
```

```python
def gradient_descent(X, Y, theta, alpha, iteration):
  '''
  X : 2D array training samples
  Y : 1D array target values
  Theta: parameters
  Alpha: learning rate
  Iterations : number of iterations

  returns:
  Theta: updated parameters
  cost_history: cost from each iteration
  '''
  cost_history = np.zeros(iteration)

  for i in range(iteration):
    # perform gradient descent formula
    #
    predictions = X.dot(theta)
    errors = np.subtract(predictions, Y)
    sum_delta = (alpha / m) * X.transpose().dot(errors)
    theta -= sum_delta
    cost_history[i] = compute_cost(X, Y, theta)

  return theta, cost_history
```

```python
iteration = 1500
alpha = 0.01

## X1
theta1, cost_history1 = gradient_descent(X1, Y, theta1, alpha, iteration)
print("Final value of theta1 is", theta1)
print("cost_history = ", cost_history1)

## X2
theta2, cost_history2 = gradient_descent(X2, Y, theta2, alpha, iteration)
print("\n\nFinal value of theta2 is", theta2)
print("cost_history = ", cost_history2)

## X3
theta3, cost_history3 = gradient_descent(X3, Y, theta3, alpha, iteration)
print("\n\nFinal value of theta2 is", theta3)
print("cost_history = ", cost_history3)
```

```python
# Plot linear regression model for X1
plt.scatter(X1[:, 1], Y, color = 'red', marker = '+', label = "Training Data")
plt.plot(X1[:, 1], X1.dot(theta1), color = 'green', label = "Linear Regression")

plt.grid(True)
plt.title("X1 column vs Y with Linear Regression Line")
plt.show()


# Plot linear regression model for X2
plt.scatter(X2[:, 1], Y, color = 'red', marker = '+', label = "Training Data")
plt.plot(X2[:, 1], X2.dot(theta1), color = 'green', label = "Linear Regression")

plt.grid(True)
plt.title("X2 column vs Y with Linear Regression Line")
plt.show()


# Plot linear regression model for X3
plt.scatter(X3[:, 1], Y, color = 'red', marker = '+', label = "Training Data")
plt.plot(X3[:, 1], X3.dot(theta1), color = 'green', label = "Linear Regression")

plt.grid(True)
plt.title("X3 column vs Y with Linear Regression Line")
plt.show()

```
